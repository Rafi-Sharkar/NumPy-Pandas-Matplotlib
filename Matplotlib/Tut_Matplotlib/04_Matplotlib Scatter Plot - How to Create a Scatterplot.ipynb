# MatPlotlib Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [2,3,4,5,6,8]
y = [2,3,7,3,4,5]
z = [5,4,2,6,7,9]

col = ['r','y','g','b','r','g']
col1 = [10,49,30,29,56,20]
siz = [400, 200, 250, 100, 650, 40]


plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Rafi")
plt.ylabel("Alif")
plt.scatter(x,y,c=col1, s=siz, marker = '*', edgecolor = 'g',linewidth= 1,cmap='BrBG', alpha=0.7)

plt.scatter(x,z,c=col, s=siz)


t=plt.colorbar()
t.set_label('Color Bar')
plt.show()
```
