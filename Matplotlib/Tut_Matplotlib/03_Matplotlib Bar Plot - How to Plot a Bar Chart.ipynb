# Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [40, 65, 35, 55]
width = .4
p = np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel("Language",fontsize=12)
plt.ylabel("Uses",fontsize=12)
plt.title("Bar Plot", fontsize = 20)
plt.bar(p,y, width,color = 'r', label='2019')
plt.bar(p1,z, width,color = 'y', label='2023')


plt.xticks(p+width/2,x,rotation=30)
plt.legend()
plt.show()
```

```python
x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [40, 65, 35, 55]

width = .4

p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Language",fontsize=12)
plt.ylabel("Uses",fontsize=12)
plt.title("Bar Plot", fontsize = 20)

plt.barh(p,y, width,color = 'r', label='2019')
plt.barh(p1,z, width,color = 'y', label='2023')


plt.xticks(p+width/2,x,rotation=30)
plt.legend()
plt.show()
```
